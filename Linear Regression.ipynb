```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4])
y = np.array([3,5,7,9])
```

```python
def cost(y_pred, y_true):
    cost_val = (np.mean((y_pred - y_true)**2))/2
    return cost_val
```

```python
w = 0.01
b = 0.02
lr = 0.01
itr = 100000
```

```python
costs = []
for i in range(itr):
    y_pred = np.dot(w,x) + b
    costs.append(cost(y_pred, y))
    diff_val = y_pred - y
    del_w = (np.dot(diff_val, x))/x.shape[0]
    del_b = np.mean(diff_val)
    w = w - lr * del_w
    b = b - lr * del_b
```

```python
y_pred = np.dot(w,x) + b
plt.plot(y_pred)
plt.scatter(x,y)
```

```python
print("w = ",w)
print("b = ",b)
```
