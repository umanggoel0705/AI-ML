```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2,3,4])
y_train = np.array([3,5,7,9])
```

```python
def cost(y_pred, y):
    return np.mean((y_pred - y)**2)/2
```

```python
itr = 10000
alpha = 0.01
w = 0.002
b = 0.003
```

```python
costs = []
for i in range(itr):
    y_pred = np.dot(w,x_train) + b
    costs.append(cost(y_pred, y_train))
    diff_val = y_pred - y_train
    del_w = np.dot(diff_val, x_train)/x_train.shape[0]
    del_b = np.mean(diff_val)
    w = w - alpha * del_w
    b = b - alpha * del_b
```

```python
plt.plot(costs)
```

```python
print(w)
print(b)
```

```python
x_test = np.array([10,11, 14, 16])
y_test = np.dot(w, x_test) + b
```

```python
plt.scatter(x_test, y_test, c='r')
plt.title("test graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
