```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2,5,6])
y_train = np.array([0,0,1,1])
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def Loss(y_pred, y):
    return (-np.dot(y, np.log(y_pred)) - np.dot((1-y),np.log(1-y_pred))) / y.shape[0]
```

```python
itr = 10000
alpha = 0.01
w = 0.002
b = 0.003
```

```python
costs = []
for i in range(itr):
    z = np.dot(w,x_train) + b
    y_pred = sigmoid(z)
    loss = Loss(y_pred, y_train)
    costs.append(loss)
    diff_val = y_pred - y_train
    del_w = np.dot(diff_val, x_train)/x_train.shape[0]
    del_b = np.mean(diff_val)
    w = w - alpha * del_w
    b = b - alpha * del_b
```

```python
plt.plot(costs)
```

```python
print(w)
print(b)
```

```python
# plt.plot(x_train,y_train)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('Logistic Regression')
plt.scatter(x_train,y_train)
plt.show()
```

```python
# converting probability into 1 or 0
def LogReg(y_pred):
    for i in range(y_pred.shape[0]):
        if y_pred[i] > 0.5:
            y_pred[i] = 1
        else:
            y_pred[i] = 0
    return y_pred
```

```python
x_test = np.array([2,3,4,3.5])
z = np.dot(w,x_test) + b
y_test = sigmoid(z)
```

```python
y_test = LogReg(y_test)
```

```python
plt.scatter(x_test, y_test, c='g')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('Logistic Regression')
plt.scatter(x_train,y_train)
plt.show()
```
